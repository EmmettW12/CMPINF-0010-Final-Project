Specifically, we used crime rate, median age of death, and fire incidents as our metrics. I worked with a data set of fire incidents to determine which neighborhood has the highest rate of fire incidents.

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "C:\\Users\\ewill\\Downloads\\FireIncidents.csv"
df = pd.read_csv(file_path)

#group data for plotting
nbh_group = df.groupby('neighborhood')
incident_counts = nbh_group.size().reset_index(name='incident_count')
incident_counts = incident_counts.sort_values(by='incident_count', ascending=False)
iC_top = incident_counts.head(20)


#format the chart 
plt.figure(figsize=(12, 6))
plt.bar(iC_top['neighborhood'], iC_top['incident_count'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title('Incident Counts by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Incident Count')
plt.tight_layout()

plt.show()
```

According to this dataset, the neighborhoods with the most incidents are Carrick, Central Business  District, South Side Flats, Squirrel Hill South, and Hazelwood.
